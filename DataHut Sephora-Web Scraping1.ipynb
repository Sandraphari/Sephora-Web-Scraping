```python
# Scrolling down the page in order to overcome Lazy Loading
time.sleep(3)

element=driver.find_element_by_tag_name('body')
while True:
    element.send_keys(Keys.PAGE_DOWN)
    time.sleep(3)
```

```python
# Converting the content of the page to BeautifulSoup object
content = driver.page_source
soup = BeautifulSoup(content,'html5lib')
```

```python
# Fetching the product links of all items 
links = []
main = soup.find_all('div',attrs={"class": "css-1322gsb"})[0]
for row1 in main.find_all('div', {'class':'css-foh208'}):
    for row2 in row1.find_all('a', {'class':'css-klx76'}):
        if row2['href'].startswith('https:'):
            links.append(row2['href'])
        else:
            links.append('https://www.sephora.com'+row2['href'])
print(links)
```

```python
# Fetching the product links- Continuation (Lazy loaded items)
for row1 in main.find_all('div', {'class':'css-1qe8tjm'}):
    for row2 in row1.find_all('a', {'class':'css-klx76'}):
        if row2['href'].startswith('https:'):
            links.append(row2['href'])
        else:
            links.append('https://www.sephora.com'+row2['href'])
print(links)
```

```python
# Creating a dictionary of the required columns
data_dic = {'product_url': [], 'brand': [], 'product_name': [],
                'number_of_reviews': [], 'love_count': [], 'star_rating': [], 'price':[], 'Fragrance Family':[],'Scent Type':[],'Key Notes':[],'Fragrance Description':[],'COMPOSITION':[],'Ingredients':[]}
```

```python
# Creating a dataframe with those columns
data = pd.DataFrame(data_dic)
```

```python
# Assigning the scraped links to the column 'product_url'
data['product_url'] = links
```

```python
# Scraping data like 'brand_name','product_name','number_of_reviews','love_count','star_rating' and 'price'
for i in range(len(data)):
    driver.get(data['product_url'].iloc[i])
    content = driver.page_source
    soup = BeautifulSoup(content,'html5lib')
    
    try:
        # brands
        brand = soup.find('a',attrs={"data-at": "brand_name"}).text
        data['brand'].iloc[i] = brand

        # product_name
        product = soup.find('span',attrs={"data-at": "product_name"}).text
        data['product_name'].iloc[i] = product
        
        # number_of_reviews
        reviews = soup.find('span',attrs={"data-at": "number_of_reviews"}).text
        data['number_of_reviews'].iloc[i] = reviews

        # love_count
        love = soup.find('span',attrs={"class": "css-jk94q9"}).text
        data['love_count'].iloc[i] = love
        
        # star_rating
        star = soup.find('span',attrs={"class": "css-1tbjoxk"})['aria-label']
        data['star_rating'].iloc[i] = star

        # price
        price = soup.find('b',attrs={"class": "css-0"}).text
        data['price'].iloc[i] = price
        
        
    except:
        pass
```
